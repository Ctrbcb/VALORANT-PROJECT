```python
import pandas as pd

agents = pd.read_csv('agents_abilities.csv')
agents_stat = pd.read_csv('valo_agents_stat.csv')

agents.head()
```

```python
agents_stat.head()
```

```python
agents_stat = agents_stat.drop(agents_stat[(agents_stat["Game Type"] != "Competitive")].index, axis = 0)
agents_stat
```

```python
agents_stat["KD"].astype(float)
agents_stat["Win %"] = agents_stat["Win %"].str.rstrip('%').astype('float')
agents_stat["Pick %"] = agents_stat["Pick %"].str.rstrip('%').astype('float')
agents_stat["First Blood %"] = agents_stat["First Blood %"].str.rstrip('%').astype('float')
agents_stat["Matches"] = agents_stat["Matches"].str.replace(',', '').astype('int')
agents_stat["Avg. Score"] = agents_stat["Avg. Score"].astype('int')
```

```python
agents_stat["KDA"] = agents_stat["KDA"].str.replace(" ","")
agents_stat["KDA"] = agents_stat["KDA"].apply(lambda x: x.split("/"))

agents_stat
```

```python
agents_stat.to_csv('preprocessed_stats.csv', index=False)
```
